Using Mean Arterial Pressure
MAP = DP + 1/3 (SP - DP)

```python
def find_MAP(bp):
  SP,DP = bp.split('/')
  SP = int(SP)
  DP = int(DP)
  MAP = DP + (SP-DP)/3
  return MAP
```

```python
df['Blood Pressure'] = df['Blood Pressure'].map(find_MAP)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_train_continous = X_train[['Age','Sleep Duration','Quality of Sleep','Physical Activity Level','Stress Level','Blood Pressure','Heart Rate','Daily Steps']]
X_test_continous = X_test[['Age','Sleep Duration','Quality of Sleep','Physical Activity Level','Stress Level','Blood Pressure','Heart Rate','Daily Steps']]
X_train_categorical = X_train[['BMI Category','Gender']]
X_test_categorical = X_test[['BMI Category','Gender']]
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

Standard Scaling

```python
X_train_continous_scaled = scaler.fit_transform(X_train_continous)
X_test_continuos_scaled = scaler.transform(X_test_continous)
```

```python
df1=pd.DataFrame(X_train_continous_scaled,columns =['Age','Sleep Duration','Quality of Sleep','Physical Activity Level','Stress Level','Blood Pressure','Heart Rate','Daily Steps'])
df2 =pd.DataFrame(X_train_categorical,columns = ['BMI Category','Gender'])
df1.reset_index(inplace=True)
df2.reset_index(inplace=True)
X_train_final = pd.concat([df1,df2],axis=1)
```

```python
df1=pd.DataFrame(X_test_continuos_scaled,columns =['Age','Sleep Duration','Quality of Sleep','Physical Activity Level','Stress Level','Blood Pressure','Heart Rate','Daily Steps'])
df2 =pd.DataFrame(X_test_categorical,columns = ['BMI Category','Gender'])
df1.reset_index(inplace=True)
df2.reset_index(inplace=True)
X_test_final = pd.concat([df1,df2],axis=1)
```

```python
X_train_final.drop('index',axis=1,inplace=True)
X_test_final.drop('index',axis=1,inplace=True)
```

```python
def tune_clf_hyperparameters(clf,param_grid,X_train,y_train,scoring='accuracy',n_splits=3):

  cv = StratifiedKFold(n_splits=n_splits,shuffle=True,random_state=0)
  clf_grid = GridSearchCV(clf,param_grid,cv=cv,scoring=scoring,n_jobs=-1)

  clf_grid.fit(X_train,y_train)

  best_hyperparameters = clf_grid.best_params_

  return clf_grid.best_estimator_,best_hyperparameters
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV, StratifiedKFold
from sklearn.metrics import classification_report, accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
```

```python
def evaluate_model(model,X_test,y_test,model_name):
  y_pred = model.predict(X_test)

  report = classification_report(y_test, y_pred, output_dict=True)

  metrics = {
        "precision_0": report["0"]["precision"],
        "precision_1": report["1"]["precision"],
        "recall_0": report["0"]["recall"],
        "recall_1": report["1"]["recall"],
        "f1_0": report["0"]["f1-score"],
        "f1_1": report["1"]["f1-score"],
        "precision_2": report["2"]["precision"],
        "recall_2": report["2"]["recall"],
        "f1_2": report["1"]["f1-score"],
        "macro_avg_precision": report["macro avg"]["precision"],
        "macro_avg_recall": report["macro avg"]["recall"],
        "macro_avg_f1": report["macro avg"]["f1-score"],
        "accuracy": accuracy_score(y_test, y_pred)
    }

  # Convert dictionary to dataframe
  df = pd.DataFrame(metrics, index=[model_name]).round(2)

  return df
```

```python
dt_evaluation = evaluate_model(best_dt, X_test_final, y_test, 'DT')
dt_evaluation
```
